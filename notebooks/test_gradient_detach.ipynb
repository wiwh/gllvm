```python
import torch

W = torch.tensor([[2.0]], requires_grad=True)
z = torch.tensor([[1.0]], requires_grad=True)
y = torch.tensor([[3.0]])

def loss_fn(W, z):
    return (W @ z - y).pow(2).sum()

loss = loss_fn(W, z)
loss.backward()

print("dL/dz =", z.grad)
print("dL/dW =", W.grad)
```

```python
z = torch.tensor([[1.0]], requires_grad=True)
loss = loss_fn(W.detach(), z)  # <-- detach W
loss.backward()

print("dL/dz (detached) =", z.grad)
```
